```python
# Section C – Short Coding Questions

# 16)  using a for loop calculate a sum of n natural numbers

n = int(input("Enter n : "))
sum = 0
for i in range(1, n + 1) :
    sum += i
print("Sum of digits are : ",sum)
```

```python
# 17)  Using a while loop count how many digits are present in a given number

num = int(input("Enter number : "))
count = 0
while num > 0 :
    count += 1
    num //= 10
print("The number of digits : ",count)
```

```python
#  18) Using a for loop find factorial of a number

n = int(input("Enter number : "))
fact = 1
for i in range(1, n + 1) :
    fact *= i
print("Factorial : ",fact)
```

```python
#  19) Using a while loop reverse a number (example : 123 → 321)

num = int(input("Enter number : "))
rev = 0
while num > 0 :
    rev = rev * 10 + num % 10
    num //= 10
print("The reverse of of number : ",rev)
```

```python
#  20) Using loops find the largest number among n user inputs

n = int(input("Enter count : "))
largest = int(input("Enter number : "))
for i in range(n - 1) :
    num = int(input("Enter number : "))
    if num > largest :
        largest = num
print("The number is largest in all numbers : ",largest)
```

```python
#  Section D – Long Coding Questions

# 21) Write a program to check whether a number is prime or not using loops only

num = int(input("Enter number : "))
if num > 1 :
    for i in range(2, num) :
        if num % i == 0 :
            print("number is not prime")
            break
    else :
        print("number is prime")
else :
    print("number is not prime")
```

```python
# 22) Write a program to generate the Fibonacci series up to n terms using a loop

n = int(input("Enter terms : "))
a = 0
b = 1

for i in range(n) :
    print(a, end=" ")
    a, b = b, a + b
```
